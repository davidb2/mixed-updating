```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from bitarray import bitarray
from utils import yield_all_graph6, yield_all_digraph6
from pathlib import Path

import moran
```

```python
N = 3
r = 1.1
# G = nx.wheel_graph(N)
# len(G)
```

```python
fp_wm = (1-1/r)/(1-1/r**N)
for G in yield_all_graph6(Path(f'./data/connected-n{N}.g6')):
# for G in yield_all_digraph6(Path(f'./data/direct{N}.d6')):
  # T = lambda v: sum(1/G.degree(u) for u in G.neighbors(v))/N
  # temps = np.array([T(v) for v in G.nodes()])
  temps = np.array([1.]*N)
  temps /= temps.sum()
  S = bitarray([0]*(N-1)+[1])
  single_mutant_locations = [
    moran.vec_2_idx(list(S << k))
    for k in range(N)
  ]
  P = np.array([
    moran.mixed_moran_fix_prob_sys(G, r=r, p_Bd=1-delta)[single_mutant_locations]
    for delta in np.linspace(0, 1, 10+1)
  ]).T
  fp_G = (P.mean(axis=1) * temps).sum()
  if True or fp_G >= fp_wm:
    print(fp_wm, fp_G)#, nx.to_digraph6_bytes(G))
```
